```python
import pandas as pd
import nlpaug.augmenter.word as naw
import fairseq
import fastBPE
import sacremoses
import transformers
from itranslate import itranslate
import time
```

In csv format

```python
def translate(text):
  try:
    translated = itranslate(text, to_lang='fil', from_lang='en')
  except:
    print(text)
    print()
    return ""

  return translated
```

```python
filename = 'wikipedia_manual_qac.csv'
df = pd.read_csv(filename)
df['Context'] = df.apply(lambda row : translate(row['Context']), axis=1)
df['Question'] = df.apply(lambda row : translate(row['Question']), axis=1)
df['Answer'] = df.apply(lambda row : translate(row['Answer']), axis=1)
df.drop(columns=['start', 'end'], inplace=True)
df
```

```python
def get_start_loc(context, answer):
  index = context.find(answer)
  if index == -1:
    return ""
  else:
    return index

def get_end_loc(context, answer, start):
  if start == "":
    return ""
  else:
    if start + len(answer) > len(context):
      return len(context)
    else:
      return start + len(answer)
```

```python
df['start'] = df.apply(lambda row : get_start_loc(row['Context'], row['Answer']), axis=1)
df['end'] = df.apply(lambda row : get_end_loc(row['Context'], row['Answer'], row['start']), axis=1)
```

```python
df_new = df.loc[(df['Context'] != "") & (df['Question'] != "") & (df['Answer'] != "") & (df['start'] != "") & (df['end'] != "")]
```

```python
df_new.to_csv('wiki_manual_tagalog_translated.csv')
```
