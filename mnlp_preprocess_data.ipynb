```python
def get_start_pos(context, answer):
  return context.find(answer)
```

```python
def get_end(start, answer):
  return start + len(answer)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('qa_120.csv')
df['start'] = df.apply(lambda row: get_start_pos(row['context'], row['answer']), axis=1)

df = df[df.start > -1]

df['end'] = df.apply(lambda row: get_end(row['start'], row['answer']), axis=1)

df.to_csv('qa_onetwenty.csv')
```
