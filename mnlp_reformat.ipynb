```python
import pandas as pd
```

```python
all_results = list()

with open('/content/drive/MyDrive/mnlp/final/spanish_squad_to_tagalog.txt', 'r') as filehandle:
  content = filehandle.readlines()
  for c in content:
    inner_dict = dict()
    parts = c.split("||")
    inner_dict['Context'] = parts[0]
    inner_dict['Question'] = parts[1]
    inner_dict['Answer'] = parts[2]
    inner_dict['start'] = inner_dict['Context'].find(inner_dict['Answer'])
    inner_dict['end'] = inner_dict['start'] + len(inner_dict['Answer'])

    all_results.append(inner_dict)

df = pd.DataFrame(all_results)
df.to_csv('/content/drive/MyDrive/mnlp/final/spanish_to_tagalog_squad.csv', index=None)
```
