```python
import json
from itranslate import itranslate
import pandas as pd
import time
```

```python
squad_tagalog = dict()
squad_tagalog['version'] = "v2.0"
squad_tagalog['data'] = list()

squad_es = None
with open('/content/train-v2.0-es.json', encoding='utf-8') as filehandle:
  squad_es = json.loads(filehandle.read())
```

```python
all_data = list()

for d in squad_es['data']:
  for p in d['paragraphs']:
    context = p['context']
    for q in p['qas']:
      question = q['question']
      for a in q['answers']:
        text = a['text']
        answer_start = a['answer_start']

        try:
          translated_context = itranslate(context, to_lang='fil', from_lang='es')
          time.sleep(0.5)
          translated_question = itranslate(question, to_lang='fil', from_lang='es')
          time.sleep(0.5)
          translated_answer = itranslate(text, to_lang='fil', from_lang='es')
          time.sleep(0.5)

          answer_start = translated_context.find(translated_answer)
          if answer_start == -1:
            continue
          else:
            inner_dict = dict()
            inner_dict['context'] = translated_context
            inner_dict['question'] = translated_question
            inner_dict['answer'] = translated_answer
            inner_dict['answer_start'] = answer_start
            with open('/content/drive/MyDrive/mnlp/final/tagalog_squad_train_csv.csv', 'a') as filehandle:
              
            all_data.append(inner_dict)

        except:
          continue

df = pd.DataFrame(all_data)
df.to_csv('/content/drive/MyDrive/mnlp/final/tagalog_squad_train_csv.csv')
```
